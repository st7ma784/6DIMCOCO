```python
#open results.pkl
import pickle
import torch
import numpy as np
import matplotlib.pyplot as plt

with open('results.pkl', 'rb') as f:
    results = pickle.load(f)
for key in results.keys(): # for how many dimensions
    #print(key)
    print(results[key])
    ##print('\n')
    #MAKE NUMPY ARRAY OF RESULTS
    #score for each method.
    y = torch.stack([ v["score"] if v is not None else torch.zeros([]) for v in results[key].values()]).numpy()
    #plot
    x = np.arange(0, len(y))
    #scale x axis exponentially
    y = np.exp(y)
    plt.plot(x, y, label=key)
    
    #label axes
    plt.xlabel('Method')
    plt.ylabel('Score')

#show plot
plt.legend()
plt.show()
```
