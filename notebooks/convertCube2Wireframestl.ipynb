```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_cuboid(size, center=None, ax=None, **kwargs):
    """
       Create a data array for cuboid plotting.


       ============= ================================================
       Argument      Description
       ============= ================================================
       center        center of the cuboid, triple
       size          size of the cuboid, triple, (x_length,y_width,z_height)
       :type size: tuple, numpy.array, list
       :param size: size of the cuboid, triple, (x_length,y_width,z_height)
       :type center: tuple, numpy.array, list
       :param center: center of the cuboid, triple, (x,y,z)
   """
    # suppose axis direction: x: to left; y: to inside; z: to upper
    # get the (left, outside, bottom) point
    import numpy as np
    if center is None:
        center=[s/2 for s in size]
    ox, oy, oz = center
    l, w, h = size

    x = np.linspace(ox-l/2,ox+l/2,num=l)
    y = np.linspace(oy-w/2,oy+w/2,num=w)
    z = np.linspace(oz-h/2,oz+h/2,num=h)
    x1, z1 = np.meshgrid(x, z)
    y11 = np.ones_like(x1)*(oy-w/2)
    y12 = np.ones_like(x1)*(oy+w/2)
    x2, y2 = np.meshgrid(x, y)
    z21 = np.ones_like(x2)*(oz-h/2)
    z22 = np.ones_like(x2)*(oz+h/2)
    y3, z3 = np.meshgrid(y, z)
    x31 = np.ones_like(y3)*(ox-l/2)
    x32 = np.ones_like(y3)*(ox+l/2)

    import matplotlib.pyplot as plt
    if ax is None:
        fig = plt.figure()
        ax = fig.add_subplot(projection='3d')
    # outside surface
    print(x1.shape)
    ax.plot_wireframe(x1, y11, z1, color='b', rstride=1, cstride=1, alpha=0.6)
    # inside surface
    ax.plot_wireframe(x1, y12, z1, color='b', rstride=1, cstride=1, alpha=0.6)
    # bottom surface
    ax.plot_wireframe(x2, y2, z21, color='b', rstride=1, cstride=1, alpha=0.6)
    # upper surface
    ax.plot_wireframe(x2, y2, z22, color='b', rstride=1, cstride=1, alpha=0.6)
    # left surface
    ax.plot_wireframe(x31, y3, z3, color='b', rstride=1, cstride=1, alpha=0.6)
    # right surface
    ax.plot_wireframe(x32, y3, z3, color='b', rstride=1, cstride=1, alpha=0.6)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    return ax


def test():
    center = [5, 5, 5]
    
    ax= plot_cuboid((10, 10, 10))
    

test()
```

To make a mesh we need a list of vertexes, and a list of faces.

The vertexes are simple for a wire mesh, they are just a list of points in 3D space. The faces are a list of indexes into the vertex list, which define the edges of the faces.

the faces are a list of vertexes in a counter-clockwise order. The order is important, as it defines the direction of the face normal. The normal is used to determine which side of the face is the front, and which is the back. The front side is the side that is visible, the back side is the side that is hidden.

This means that for every cube we need 8 vertexes and 12 !!! faces. each face is a triangle so 2 are needed for each side of the cube.

```python
def find_corners(cube_top_left):
    #print(cube_top_left)
    x,y,z=cube_top_left
    return [
    (x,y,z),
    (x+1,y,z),
    (x,y+1,z),
    (x,y,z+1),
    (x+1,y+1,z),
    (x+1,y,z+1),
    (x,y+1,z+1),
    (x+1,y+1,z+1)
     ]
def make_sides(corners_coordinates):
    return [
        (corners_coordinates[7],corners_coordinates[4],corners_coordinates[2]),
        (corners_coordinates[2],corners_coordinates[6],corners_coordinates[7]),

        (corners_coordinates[3],corners_coordinates[1],corners_coordinates[5]),
        (corners_coordinates[3],corners_coordinates[0],corners_coordinates[1]),

        (corners_coordinates[1],corners_coordinates[2],corners_coordinates[4]),
        (corners_coordinates[0],corners_coordinates[2],corners_coordinates[1]),

        (corners_coordinates[1],corners_coordinates[7],corners_coordinates[5]),
        (corners_coordinates[1],corners_coordinates[4],corners_coordinates[7]),

        (corners_coordinates[5],corners_coordinates[7],corners_coordinates[6]),
        (corners_coordinates[5],corners_coordinates[6],corners_coordinates[3]),

        (corners_coordinates[3],corners_coordinates[6],corners_coordinates[0]),
        (corners_coordinates[0],corners_coordinates[6],corners_coordinates[2])

    ]

def make_wireframe(shape,step=1):
    #this will be a cube, so 8 points
    cube_x,cube_y,cube_z=shape
    #we need 3 sets of lines,
    #1. lines of length 'cube_x' parallel to x axis, for each y,z
    #2. lines of length 'cube_y' parallel to y axis, for each x,z
    #3. lines of length 'cube_z' parallel to z axis, for each x,y
    y_range=np.arange(cube_y,step=step)
    z_range=np.arange(cube_z,step=step)
    x_range=np.arange(cube_x,step=step)
    offset=0.1
    #the first item in each range to offset by 0.4
    y_range[0]+=offset
    z_range[0]+=offset
    x_range[0]+=offset

    #1 parallel to x axis for each y,z
    gridz,gridy=np.meshgrid(y_range,z_range)
    lines=[]
    #create a tuple of start stop coordinates for each pair of y,z
    for each in zip(gridy.flatten(),gridz.flatten()):
        cube_Vertices=[
        (offset,each[0]-offset,each[1]-offset),
        (cube_x,each[0]-offset,each[1]-offset),
        (offset,each[0]+offset,each[1]-offset),
        (offset,each[0]-offset,each[1]+offset),
        (cube_x,each[0]+offset,each[1]-offset),
        (cube_x,each[0]-offset,each[1]+offset),
        (offset,each[0]+offset,each[1]+offset),
        (cube_x,each[0]+offset,each[1]+offset)
        ]
        '''   
        (x,y,z),
        (x+1,y,z),
        (x,y+1,z),
        (x,y,z+1),
        (x+1,y+1,z),
        (x+1,y,z+1),
        (x,y+1,z+1),
        (x+1,y+1,z+1)
        ]'''
        #make all faces 
        lines.extend(make_sides(cube_Vertices))
    #2 parallel to y axis for each x,z
    gridz,gridx=np.meshgrid(z_range,x_range)
    for each in zip(gridx.flatten(),gridz.flatten()):
        cube_Vertices=[
        (each[0]-offset,offset,each[1]-offset),
        (each[0]+offset,offset,each[1]-offset),
        (each[0]-offset,cube_y,each[1]-offset),
        (each[0]-offset,offset,each[1]+offset),
        (each[0]+offset,cube_y,each[1]-offset),
        (each[0]+offset,offset,each[1]+offset),
        (each[0]-offset,cube_y,each[1]+offset),
        (each[0]+offset,cube_y,each[1]+offset)]
        '''(x,y,z),
           (x+1,y,z),
           (x,y+1,z),
           (x,y,z+1),
           (x+1,y+1,z),
           (x+1,y,z+1),
           (x,y+1,z+1),
           (x+1,y+1,z+1)
        ]'''
        lines.extend(make_sides(cube_Vertices))
    
    #3 parallel to z axis for each x,y
    gridx,gridy=np.meshgrid(x_range,y_range)
    for each in zip(gridx.flatten(),gridy.flatten()):
        cube_Vertices=[
        (each[0]-offset,each[1]-offset,offset),
        (each[0]+offset,each[1]-offset,offset),
        (each[0]-offset,each[1]-offset,cube_z),
        (each[0]-offset,each[1]+offset,offset),
        (each[0]+offset,each[1]-offset,cube_z),
        (each[0]+offset,each[1]-offset,cube_z),
        (each[0]-offset,each[1]+offset,cube_z),
        (each[0]+offset,each[1]+offset,cube_z)]

        lines.extend(make_sides(cube_Vertices))
    return lines



def make_faces(cube):
    x,y,z=np.nonzero(cube)

    #make a list of xyz coordinates of each cube 
    cubes=list(zip(x,y,z))
    faces=[]
    for each in cubes:
        corners_coordinates=find_corners(each)
        faces.extend(make_sides(corners_coordinates))
    
    #next add a wire frame to the cube


    faces.extend(make_wireframe(cube.shape))
    #every item in faces is a set of 3 points, each point is a tuple of coordinates. 
    #we're going to make a list of points.
    points=[]
    for each in faces:
        points.extend(each)
    #now we have a list of points, we need to remove duplicates
    points=list(set(points))
    #now we need to assign each point an index
    
    #now we need to replace each point in faces with its index
    #first we need to create a dictionary of points and their indices
    point_index_dict={}
    for index,point in enumerate(points):
        point_index_dict[point]=index
    #now we need to replace each point in faces with its index
    new_faces=[]
    for face in faces:
        #face is a tuple of 3 points 
        new_face=[]
        for point in face:
            new_face.append(point_index_dict[point])
        new_faces.append(new_face)


    return points,new_faces





# indices_list=[]
# vertexes,faces=make_faces(cube)

# #convert to a mesh object
# from stl import mesh
# import numpy as np
# # Create the mesh
# cube_mesh = mesh.Mesh(np.zeros(len(faces), dtype=mesh.Mesh.dtype))
# for i, f in enumerate(faces):
#     for j in range(3):
#         cube_mesh.vectors[i][j] = vertexes[f[j]][:]

# # Write the mesh to file "cube.stl"
# cube_mesh.save('cube6.stl')
```
